import pandas as pd
import matplotlib.pyplot as plt

# 1. Data loading and cleaning

df = pd.read_csv('C:/Users/ARYA/Downloads/ecommerce/Amazon Sale Report.csv',low_memory = False)



```python
df.head()
```

```python
df = df.drop(['promotion-ids', 'fulfilled-by', 'Unnamed: 22'], axis=1)
```

```python
df = df.dropna(subset=['Amount','currency'])
```

```python
df[['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']] = df[[
    'ship-city', 'ship-state', 'ship-postal-code', 'ship-country'
]].fillna('Unknown')


df['Courier Status'] = df['Courier Status'].fillna("Pending")
```

# 3. Visualization and Insights

```python
Total_sales = df['Amount'].sum()
print("Total_sales------>",Total_sales)

sales_by_country = df.groupby('ship-country')['Amount'].sum().sort_values(ascending=False)
print("Sales by Country:")
print(sales_by_country)

Total_b2b = df['B2B'].sum()
print("Total_b2b------>",Total_b2b)
```

```python
#pie chart of b2b orders
b2b_counts = df['B2B'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(
    b2b_counts, 
    labels=['Non-B2B', 'B2B'],  # Order matters: False first, True second
    autopct='%1.1f%%', 
    startangle=45,
    colors = ['#76d7c4','#27ae60']
)

plt.title('B2B vs Non-B2B Orders')
plt.axis('equal')  # Makes the pie a perfect circle
plt.tight_layout()
plt.show()
```

```python
# Plot bar chart

# Group sales by country
state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
state_sales = state_sales.head(15)



plt.figure(figsize=(10,6))
state_sales.plot(kind='bar', color='mediumseagreen')
plt.title('Total Sales by states')
plt.xlabel('State')
plt.ylabel('Revenue (â‚¹)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#monthly sales line chart
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

monthly_sales = df.resample('m', on='Date')['Amount'].sum()


plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', color='darkgreen', marker='o')  # or monthly_sales
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Revenue (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# key insights

- B2B orders represent only 99.3% of total transactions
- Maharashtra, Karnataka, and telangana are the top-performing states
- Sales spiked during April
